```python
import numpy as np
import torch
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageOps

```

```python
#create image object color and grayscale using pillow library
orig = Image.open('images/panda.JPG')
pic = np.array(orig)
grayimg = ImageOps.grayscale(orig)
grayimg.save('images/gray_panda.JPG')
gray = np.array(grayimg)
#pic_tensor = torch.from_numpy(pic)
#pic_tensor.size()
```

```python
plt.imshow(pic)
```

```python
plt.imshow(gray)
gray.shape
```

```python
plt.imshow(pic[:,:,0]);
plt.colorbar()
```

```python
a = np.zeros(gray.shape)
threshold = 200
```

```python
for i in range(gray.shape[0]):
    for j in range(gray.shape[1]):
        pixel = gray[i,j]
        if pixel > threshold:
            a[i,j]=gray[i,j]
        
```
